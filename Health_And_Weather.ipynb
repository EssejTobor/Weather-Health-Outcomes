```python
import pandas as pd
import numpy as np
import polypy
import matplotlib.pyplot as plt


```

```python
cities = "c:\\Resources\\Project\\10_City_update.csv"
cities_data_df = pd.read_csv(cities)
cities_data_df.head()
```

```python
weather = "c:\\Resources\\Project\\AllDf1.csv"
weather_data_df = pd.read_csv(weather)
weather_data_df.head()
```

```python
city_weather_merge_df = pd.merge(cities_data_df,weather_data_df, on=("CityName","Year"), suffixes=("_Medical","_Weather"))
city_weather_merge_df
```

```python
year = city_weather_merge_df["Year"]
city = city_weather_merge_df["CityName"]
measure = city_weather_merge_df["Data_Value"]
measure_id = city_weather_merge_df["MeasureId"]
category = city_weather_merge_df["Measure"]
data_value = city_weather_merge_df["Data_Value"]
population = city_weather_merge_df["PopulationCount"]
new_population = round(population / 10000)
snow = city_weather_merge_df["Snow 24h"]
temp = city_weather_merge_df["Temperature"]
clouds = city_weather_merge_df["Clouds All"]


summary_df = pd.DataFrame({
    "Year" : year,
    "City" : city,
    "Health value" :data_value,
    "Health Category" :measure_id,
    "Category" : measure,
    "Population": new_population,
    "Temperature" : temp,
    "Snow" : snow,
    "Cloudiness" : clouds
  
})
summary_df.head

#summary_df.to_csv("c:\\Resources\\project\\summary_health_weather_census.csv", encoding="utf-8", index=False)
```
